## open()
https://docs.python.org/3/library/functions.html#open

````python
path_txt = list_files[0]
````

````python
f = open(path_txt, 'r')
txt = f.read() # readlines()
f.close()

print(txt)
````

````python
with open(path_txt, 'r') as f:
    txt2 = f.read()

print(txt2)
````

````python
nlines = 5
txt3 = ''
with open(path_txt, 'r') as f:
    for line in range(nlines):
        txt3 += f.readline()

print(txt3)
````
