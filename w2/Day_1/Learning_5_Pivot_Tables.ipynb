### Let's find the answer to a really important question:
_"How many clowns can you fit into different types of cars?"_


```python
#We are lucky enough to have a .csv-file with the results from a groundbreaking study!
results = pd.read_csv(os.path.join(os.pardir, 'resources', 'clowns_in_cars.csv')) 

results
```

```python
import numpy as np

data = pd.read_csv(os.path.join(os.pardir, 'resources', 'clowns_in_cars.csv')) 

results = np.round(pd.pivot_table(data, values='num_clown', 
                                index=['brand','model','num_of_doors'], 
                                columns='type_clown'))

results
```

### We also want to know what type of car is the best for moving all participants of a clown convention
_Ideally we want to move all the clowns in as few cars as possible_

```python
results = pd.pivot_table(data, values='num_clown', 
                                index=['brand','model','car_weight','num_of_doors'])
results
```

From this pivot table we can then see that the Ford Kuga is the best available car for transporting the clowns,
with the possibility of transporting a mean of 18 clowns (excluding the possibility of transporting only a clown's legs in one car)
